```python
pages = []
htmls = []
for article in headlines['articles'][10:20]:
    url = article['url']
    if url != "https://removed.com":
        html = requests.get(url)
        htmls.append(html)
        page_content = BeautifulSoup(html.content)
        pages.append(page_content)
```

```python
import re

pattern = re.compile(r"_ad|-ad|advert|adwrap|^ad-|^ad_", re.IGNORECASE)

ads = []
for page, html in zip(pages, htmls):
    if html.status_code == 403:
        ads.append("Blocked")
    else:
        ad = page.find_all('div', class_=lambda c: c and pattern.search(c))
        ads.append(ad)

# Possible status codes
# 200 - ok
# 400 - bad request 
# 401 - unauthorised (requires valid credentials)
# 403 - forbidden (request understood but wont be fulfilled)
# 404 - not founce
# 429 - rate limit excessed
```

```python
for x in ads: 
    if x == "Blocked":
        print(x)
    else:
        print(len(x)) 
```

```python
# Count script hiding ads check
script_counts = []
for p in pages:
    counter = 0
    for s in p.find_all('script'):
        s_text = s.get_text()
        if s_text.lower().count("hidead") > 0 and s_text.lower().count("div") > 1:
            counter += 1

    script_counts.append(counter)

print(script_counts)
        
```
